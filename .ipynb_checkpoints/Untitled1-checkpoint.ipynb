```python
import pandas as pd
import numpy as np

pd.concat
```

```python
df1 = pd.DataFrame(np.random.rand(3, 2), columns = ['cat', 'dog'])
df2 = pd.DataFrame(np.random.rand(3, 2), columns = ['cat', 'dog'])

print(df1); print(df2)
```

```python
pd.concat([df1, df2])
```

```python
df3 = pd.DataFrame(np.random.rand(3, 2), columns = ['bird', 'horse'])

pd.concat([df1, df3], axis = 1)
```

```python
data4 = {'Alpaca': ['A1', 'A2', 'A3'], 'Bird': ['B1', 'B2', 'B3'], 'Camel': ['C1', 'C2', 'C3']}
df4 = pd.DataFrame(data4)
print(df4)

data5 = {'Bird': ['B3', 'B4', 'B5'], 'Camel': ['C3', 'C4', 'C5'], 'Duck': ['D3', 'D4', 'D5']}
df5 = pd.DataFrame(data5, index = [3, 4, 5])
print(df5)

print(pd.concat([df4, df5], sort = False))

print(pd.concat([df4, df5], join = 'inner'))
```

```python
pd.merge
# left - one of the df to merge
# right - other df to merge
# how - specifies the kind of merge to make
# on - the col on which to merge; needs to be in both df
# left_on - col to join on the left df
# right_on - col to join on the right df
# left_index | right_index - use index from left/right df as the join key(s)
```

```python
df1 = pd.DataFrame({'day': ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri'],
                    'weather': ['sunny', 'windy', 'rainy', 'cloudy', 'cold']})
df2 = pd.DataFrame({'day': ['Wed', 'Fri', 'Mon', 'Thurs', 'Tues'],
                    'temp': [88, 75, 62, 65, 53]})
display(df1); display(df2);

df3 = pd.merge(df1, df2)
display(df3)
# default behavior: order of rows in each df does not have to be same for merge
```

```python
df2_new = pd.DataFrame({'day_of_week': ['Wed', 'Fri', 'Mon', 'Thurs', 'Tues'],
                    'temp': [88, 75, 62, 65, 53]})
df4 = pd.merge(df1, df2_new, left_on = 'day', right_on = 'day_of_week')
display(df4)
```

```python
df5 = pd.merge(df1, df2_new, left_on = 'day', right_on = 'day_of_week').drop('day_of_week', axis = 1)
display(df5)
```

```python
df6 = pd.DataFrame({'name': ['Snowball', 'Bob', 'Tweetie', 'Fluffy'],
                   'type': ['cat', 'rat', 'bird', 'dog']})
df7 = pd.DataFrame({'name': ['Snowball', 'Bob'], 'weight': [400, 749]})

display(df6); display(df7)

# union
df8_outer = pd.merge(df6, df7, how = 'outer')
# intersection
df8_inner = pd.merge(df6, df7, how = 'inner')

display(df8_outer); display(df8_inner)
```

```python
# join method
# other - the other df you are joining
# on - the col or index on which to join the other df (default is index)
# how - similar to merge the choices are left, right, outer, inner

df1a = df1.set_index('day')
df2a = df2.set_index('day')

# display(df1a); display(df2a);

df1a.join(df2a, on = 'day')
```

```python
# tidy data: each variable is a column, each observation forms a row
# melt(): unpivot, id_vars - identifies the observation (row), value_vars - cols that will unpivot

df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                 'B': {0: 1, 1: 3, 2: 5}})
display(df)

pd.melt(df, id_vars = ['A'], value_vars = ['B'])
```

```python
# pivot(): index - col to use to create the new index, columns - values in this col will create new cols, values: column(s) to use for populating the values

df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two', 'two'],
                  'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                  'baz': [1, 2, 3, 4, 5, 6]})
display(df)

df.pivot(index = 'foo', columns = 'bar', values = 'baz')
```

```python
import seaborn as sns

flights = sns.load_dataset('flights')
flights.head(10)
```

```python
flights_pivot = flights.pivot(index = 'year', columns = 'month', values = 'passengers')

display(flights_pivot)
```
